```python
from keras.applications.vgg16 import VGG16
model = VGG16(weights='imagenet', include_top=False)
```

```python
from tensorflow.keras.utils import load_img, img_to_array
from keras.applications.vgg16 import preprocess_input
from numpy import expand_dims

def get_image_embedding(img):
    scale_percent = 10
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)
    img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

    img_array = expand_dims(img, axis=0)
    img_array = preprocess_input(img_array)
    embedding = model.predict(img_array)
    embedding = embedding.flatten()

    return embedding
```

```python
import cv2
from pathlib import Path
import numpy as np
import csv

cds = ['rozetked']


for cd in cds:
  path = Path(cd)
  for file in path.iterdir():
    if 'video' in str(file):
      video_embs = []
      print(file)
      cap = cv2.VideoCapture(str(file))
      count = 0

      while cap.isOpened():
          ret, frame = cap.read()

          if ret:
              print(f'{file} - {count}')
              video_embs.append(np.array(get_image_embedding(frame)))
              count += 30 # i.e. at 30 fps
              cap.set(cv2.CAP_PROP_POS_FRAMES, count)
          else:
              cap.release()
              break
      video_embs = np.array(video_embs)
      with open(f'{file}.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        for row in video_embs:
          print(len(row))
          writer.writerow(list(row))
```
