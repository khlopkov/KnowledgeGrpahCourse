# Создание графа, определение сущностей и атрибутов

```python
g = Graph()
graph = URIRef('https://w3id.org/d2s/graph/RickAndMorty')
path = 'https://w3id.org/d2s/graph/RickAndMorty#'
```

```python
#Classes
People = URIRef(path + 'People')
Places = URIRef(path + 'Places')
Episodes = URIRef(path + 'Episodes')

#people properties
name = URIRef(path + 'name')
place_of_origin = URIRef(path + 'places_of_origin')
located_in = URIRef(path + 'located_in')
status = URIRef(path + 'status')
species_of = URIRef(path + 'species_of')
type_of = URIRef(path + 'type_of')
gender = URIRef(path + 'gender')
image = URIRef(path + 'image')
```

```python
#places properties
place_name = URIRef(path + 'place_name')
place_type = URIRef(path + 'place_type')
dimension = URIRef(path + 'dimension')
place_url = URIRef(path + 'place_url')
creation_date = URIRef(path + 'creation_date')
```

```python
#episodes properties
episode_name = URIRef(path + 'episode_name')
episode_id = URIRef(path + 'episode_id')
air_date = URIRef(path + 'air_date')
episode_created = URIRef(path + 'episode_created')
episode_url = URIRef(path + 'episode_url')
character_in = URIRef(path + 'character_in')
```

```python
Utterance = URIRef(path + 'Utterance')

#utterance properties
text_link = URIRef(path + 'text_link')
utterance_image = URIRef(path + 'utterance_image')
said_by_character = URIRef(path + 'said_by_character')
said_in_episode = URIRef(path + 'said_in_episode')
start_time = URIRef(path + 'start_time')
end_time = URIRef(path + 'end_time')
sound_url = URIRef(path + 'soung_url')
```

```python
Text = URIRef(path + 'Text')

#text properties
source_text = URIRef(path + 'source_text')
wikidata_url = URIRef(path + 'wikidata_url')
```

```python
CharacterImage = URIRef(path + 'CharacterImage')

#image properties
segmented_character = URIRef(path + 'segmented_character')
image_utterance = URIRef(path + 'image_utterance')
mask = URIRef(path + 'mask')
```

# Добавление мест

```python
def AddPlace(id, row):
  place = URIRef(path + 'Places' + str(id))
  g.add((place, RDF.type, Places))
  g.add((place, place_name, Literal(row['Place Name'])))
  g.add((place, place_type, Literal(row['Place Type'])))
  g.add((place, dimension, Literal(row['Dimension'])))
  g.add((place, place_url, Literal(row['Place Url'])))
  g.add((place, creation_date, Literal(row['Place Created'], datatype=XSD.dateTime)))
```

```python
for index, row in places_data.iterrows():
  AddPlace(index, row)
```

# Добавление персонажей

```python
def AddPerson(row):
    person = URIRef(path + 'People' + str(row['ID']))
    g.add((person, RDF.type, People))
    g.add((person, name, Literal(row['Name'])))

    if row['Origin'] != 'unknown':
      tmp_origin = places_data.index[places_data['Place Name'] == row['Origin']][0]
      g.add((person, place_of_origin, URIRef(path + 'Places' + str(tmp_origin))))

    if row['Location'] != 'unknown':
      tmp_location = places_data.index[places_data['Place Name'] == row['Location']][0]
      g.add((person, located_in, URIRef(path + 'Places' + str(tmp_location))))

    g.add((person, status, Literal(row['Status'])))
    g.add((person, species_of, Literal(row['Species'])))
    g.add((person, type_of, Literal(str(row['Type']))))
    g.add((person, gender, Literal(row['Gender'])))
    g.add((person, image, Literal(row['ImageURL'])))

    return person

print(AddPerson(people_data.iloc[0]))
```

```python
for index, row in people_data.iterrows():
  AddPerson(row)
```

# Добавление эпизодов

```python
def AddEpisodes():
  for id in episodes_data['ID'].unique():
      sub_data = episodes_data[episodes_data['ID'] == id]
      episode = URIRef(path + 'Episodes' + str(id))
      g.add((episode, RDF.type, Episodes))
      g.add((episode, episode_id, Literal(id)))
      g.add((episode, episode_name, Literal(sub_data['Episode Name'].unique()[0])))
      g.add((episode, air_date, Literal(sub_data['Air Date'].unique()[0])))
      g.add((episode, episode_created, Literal(sub_data['Created'].unique()[0], datatype=XSD.dateTime)))
      g.add((episode, episode_url, Literal(sub_data['Episode URL'].unique()[0])))

      for index, row in sub_data.iterrows():
        ch = row['CharacterInEpisode'].split('/')[-1]
        g.add((episode, character_in, URIRef(path + 'People' + ch)))

```

# Добавление кадров

```python
import json

with open('instances_default.json', 'r') as f:
  image_df = json.load(f)
```

```python
t_category = Literal('Rick Sanchez')

gres = g.query("""PREFIX foaf: <https://w3id.org/d2s/graph/RickAndMorty#>
                SELECT ?x
                 WHERE {
                     ?x foaf:name ?name .
                     FILTER(?name = ?person)
                 } """, initBindings={'person': t_category})

for row in gres:
  print(row[0].replace('https://w3id.org/d2s/graph/RickAndMorty#People', ''))
```

```python
for row in image_df['annotations']:
  t_mask = row['segmentation']
  for cat in image_df['categories']:
    if row['category_id'] == cat['id']:
      t_category = Literal(cat['name'])
  for image in image_df['images']:
    if row['image_id'] == image['id']:
      t_image_id = image['file_name'].replace('dialogue_', '').replace('.png', '')

  gres = g.query("""PREFIX foaf: <https://w3id.org/d2s/graph/RickAndMorty#>
                SELECT ?x
                 WHERE {
                     ?x foaf:name ?name .
                     FILTER(?name = ?person)
                 } """, initBindings={'person': t_category})

  i = 0
  for row in gres:
    if i > 0:
      break
    person_link = row[0]
    person_id = row[0].replace('https://w3id.org/d2s/graph/RickAndMorty#People', '')
    i += 1


  character_image = URIRef(path + 'Character' + str(person_id) + 'Image' + str(t_image_id))
  g.add((character_image, RDF.type, CharacterImage))
  g.add((character_image, mask, Literal(t_mask)))
  g.add((character_image, image_utterance, URIRef(path + 'Utterance' + str(t_image_id))))
  g.add((character_image, segmented_character, URIRef(person_link)))
```
