# **Парсинг xml**

```python
import pandas as pd

columns = ['image_id', 'image_name', 'Person', 'Dog', 'Is_person']
df_image = pd.DataFrame(columns=columns)
```

```python
labels_xml = ['xtl','ytl','xbr','ybr']
```

```python
from matplotlib.cbook import boxplot_stats
import xml.etree.ElementTree as ET
from collections import defaultdict

tree = ET.parse('/content/annotations_.xml')
root = tree.getroot()

Person = defaultdict(list)
Dog = defaultdict(list)
for image in root.findall('image'):
  id_image = image.attrib.get('id')
  name_image = image.attrib.get('name')
  boxs = image.findall('box')
  Is_Person = False
  for i in range(len(boxs)):
    box_xtl = boxs[i].get('xtl')
    box_ytl = boxs[i].get('ytl')
    box_xbr = boxs[i].get('xbr')
    box_ybr = boxs[i].get('ybr')
    temp = {
      'xtl': box_xtl,
      'ytl': box_ytl,
      'xbr': box_xbr,
      'ybr': box_ybr,
    }
    if boxs[i].get('label')=='person':
      Person[id_image].append(temp)
      Is_Person = True
    if boxs[i].get('label')=='dog':
      Dog[id_image].append(temp)
  df_image.loc[len(df_image)] = [id_image,name_image,Person[id_image],Dog[id_image],Is_Person]
```

```python
df_image = df_image.drop(labels = ['image_id'],axis=1)
```

```python
label_name_list = [name.split('_')[0] for name in df_image['image_name']]
```

```python
df_image['breed'] = [labels_dict.get(i) for i in label_name_list]
```
