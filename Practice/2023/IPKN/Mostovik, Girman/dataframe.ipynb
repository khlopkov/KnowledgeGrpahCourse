```python
import pandas as pd
import os
```

```python
files_txt = [i for i in os.listdir('./') if i.endswith('.txt')]
```

```python
df_songs = pd.DataFrame(columns=['group_name', 'album_name', 'year', 'track'])
for i in sorted(files_txt):
    with open(i, 'r') as f:
        file = f.read()
    try:
        file = file.replace('Жаль, нет ружья!', 'Жаль - нет ружья!').replace('Жаль, нет Ружья!', 'Жаль - нет ружья!')
        file = file.replace('Та, что смотрит из пруда', 'Та что смотрит из пруда')
    except:
        pass
    data = file.split(', ')
    data_new_form = pd.DataFrame([data[:3] + [i] for i in data[3:]], columns=df_songs.columns)
    df_songs = pd.concat([df_songs, data_new_form])
```
