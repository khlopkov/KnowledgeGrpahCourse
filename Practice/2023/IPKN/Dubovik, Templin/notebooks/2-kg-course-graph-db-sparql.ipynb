```python
tournaments['Country'] = tournaments['Tournament'].apply(lambda x: get_country(x))
```

```python
go_ontology = Namespace('http://example.org/go_ontology.owl/')
schema = Namespace('http://schema.org/')
```

```python
gr_players = Graph()

for index, row in players.iterrows():
    player = URIRef(go_ontology + row['Name'])
    country = URIRef(go_ontology + str(row['Country']))
    
    gr_players.add((player, RDF.type, schema.Person))
    gr_players.add((country, RDF.type, schema.Country))
    
    gr_players.add((player, schema.representsTheCountry, country))
    
    gr_players.add((player, FOAF.name, Literal(row['Name'], datatype=XSD.string) ))
    
    if row['Date of Birth'] != '-':
        gr_players.add((player, schema.dateOfBirth, Literal(row['Date of Birth'], datatype=XSD.date) ))
        gr_players.add((player, schema.yearOfBirth, Literal(row['Year of Birth'], datatype=XSD.int) ))
    
    gr_players.add((player, schema.hasAGender, Literal(row['Sex'], datatype=XSD.string) ))
    gr_players.add((player, schema.hasARating, Literal(row['Rating'], datatype=XSD.int) ))
    gr_players.add((player, schema.totalPlayed, Literal(row['Total'], datatype=XSD.int) ))
    
    gr_players.add((country, schema.nameOfCountry, Literal(row['Country'], datatype=XSD.string)))
```

```python
gr_parties = Graph()

for index, row in parties.iterrows():
    party = URIRef(go_ontology + str(row['id_game']))    
    gr_parties.add((party, RDF.type, schema.Party))
    gr_parties.add((party, schema.partyId, Literal(row['id_game'], datatype=XSD.int) )) 
    gr_parties.add((party, schema.partyDate, Literal(row['Date'], datatype=XSD.date) ))
    
    player = URIRef(go_ontology + row['Name'])
    gr_parties.add((player, RDF.type, schema.Person))
    gr_parties.add((party, schema.isWinner, player ))
    
    player = URIRef(go_ontology + row['Opponent'])
    gr_parties.add((player, RDF.type, schema.Person))
    gr_parties.add((party, schema.isLoser, player ))
    
    if row['Tournament'] != '-':
        tournament = URIRef(go_ontology + row['Tournament'])
        gr_parties.add((party, schema.playedInTournament, tournament ))
        
    gr_parties.add((party, schema.partyInTheYear, Literal(row['Year'], datatype=XSD.int) ))
```

```python
gr_tournaments = Graph()

for index, row in tournaments.iterrows():
    tournament = URIRef(go_ontology + row['Tournament'])
    gr_tournaments.add((tournament, RDF.type, schema.Tournament))
    
    country = URIRef(go_ontology + str(row['Country']))
    gr_tournaments.add((country, RDF.type, schema.Country))
    
    gr_tournaments.add((tournament, schema.inTheCountry, country))
    
    player = URIRef(go_ontology + row['Winner'])
    gr_tournaments.add((player, RDF.type, schema.Person))
    gr_tournaments.add((tournament, schema.isChampion, player))
    
    player = URIRef(go_ontology + row['Runner-up'])
    gr_tournaments.add((player, RDF.type, schema.Person))
    gr_tournaments.add((tournament, schema.isRunnerup, player))
    
    gr_tournaments.add((tournament, schema.title, Literal(row['Tournament'], datatype=XSD.string) ))
    gr_tournaments.add((tournament, schema.inTheYear, Literal(row['Year'], datatype=XSD.int) ))
```

### Как зовут игрока с самым высоким рейтингом? Из какой он страны?

```python
knows_query = """
SELECT DISTINCT ?rating ?name ?countryname
WHERE {
    ?Person ns1:hasARating ?rating .
    ?Person foaf:name ?name .
    ?Person ns1:representsTheCountry ?Country .
    ?Country ns1:nameOfCountry ?countryname .
}
ORDER BY DESC(?rating)
LIMIT 1"""

qres = gr_res.query(knows_query)
for row in qres:
    print(f"{row.name}, {row.countryname} ({row.rating})")
```

### Кто самый титулованный игрок в мире?

```python
knows_query = """
SELECT ?Person ?name (COUNT(?name) AS ?count)
WHERE {
    ?Tournament ns1:isChampion ?Person .
    ?Tournament ns1:title ?title .
    ?Person foaf:name ?name .
}
GROUP BY ?Person
ORDER BY DESC(?count)
LIMIT 1
"""

qres = gr_res.query(knows_query)
for row in qres:
    print(f"{row.name} - выиграл в {row['count']} турнирах")
```

### В каком году родился Ли Чхан Хо (Lee Changho)?

```python
knows_query = """
SELECT ?name ?year
WHERE {
    ?Person foaf:name ?name .
    ?Person ns1:yearOfBirth ?year .
    FILTER (regex(?name,"Lee_Changho")) .
}
"""

qres = gr_res.query(knows_query)
for row in qres:
    print(f"{row.name} родился в {row.year} году")
```

### Кто самый молодой игрок, побеждавший Ли Чхан Хо (Lee Changho)?

```python
knows_query = """
SELECT DISTINCT ?name_win ?date
WHERE {
    ?Person_win foaf:name ?name_win .
    ?Party ns1:isWinner ?Person_win .
    ?Party ns1:isLoser ?Person_lose .
    ?Person_lose foaf:name ?name_lose .
    ?Person_win ns1:dateOfBirth ?date .
    FILTER (regex(?name_lose,"Lee_Changho")) .
}
ORDER BY DESC(?date)
LIMIT 1
"""

qres = gr_res.query(knows_query)
for row in qres:
    print(f"{row.name_win}, {row.date}")
```

### Количество партий по годам за последние 30 лет, сыгранных между про-игроками из Японии

```python
knows_query = """
SELECT ?Party ?year (COUNT(?year) AS ?count)
WHERE {
    ?Party ns1:partyInTheYear ?year .
    ?Party ns1:isWinner ?Person_win .
    ?Person_win ns1:representsTheCountry ?Country_win .
    ?Country_win ns1:nameOfCountry ?countryname_win .
    ?Party ns1:isLoser ?Person_lose .
    ?Person_lose ns1:representsTheCountry ?Country_lose .
    ?Country_lose ns1:nameOfCountry ?countryname_lose .
    FILTER ((?year > 2003) 
            && (regex(?countryname_win,"Japan"))
            && (regex(?countryname_lose,"Japan"))) .
}
GROUP BY ?year
ORDER BY ?year
"""

qres = gr_res.query(knows_query)
for row in qres:
    print(f"{row.year} - {row['count']}")
```
