```python
import sys
import os
import csv
import json


common_fields = ['symbol', 'fiftyDayAverage', 'twoHundredDayAverage', 'dayHigh', 'dayLow', 'currency', 'exchange', 'exchangeTimezoneName', 'exchangeTimezoneShortName']
etf_fields = ['navPrice', 'yield']
mutual_fund_fields = ['morningStarOverallRating', 'morningStarRiskRating', 'yield', 'ytdReturn']
equity_fields = ['sector', 'trailingEps', 'trailingPE']

def parse_data(tickers_list):
  rows = []

  data_length = len(tickers_list)
  for idx, value in enumerate(tickers_list):
      info = value[1].info
      
      if not 'quoteType' in info:
        continue

      keys = common_fields
      data_obj={'type': info['quoteType']}
      if info['quoteType'] == 'ETF':
        keys = keys + etf_fields
      elif info['quoteType'] == 'MUTUALFUND':
        keys = keys + mutual_fund_fields
      elif info['quoteType'] == 'EQUITY':
        keys = keys + equity_fields
      else:
          continue

      for key in keys:
        data_obj[key] = info.get(key, None)
      rows.append(data_obj)

      log_str = "parse: " + str(idx + 1) + " of " + str(data_length)
      sys.stdout.write("\r" + log_str)
      sys.stdout.flush()
  os.system('cls')

  return rows


result = parse_data(list(data.tickers.items()))

with open('data.json', 'w') as outfile:
    json.dump(result, outfile, indent=4, sort_keys=True)
```

```python
base_iri = 'http://www.semanticweb.org/anton/ontologies/2021/3/economics#'

#classes
lowAndHighPrice = URIRef(base_iri + 'LowAndHighPrice')
exchange = URIRef(base_iri + 'Exchange')
currency = URIRef(base_iri + 'Currency')
financialAsset = URIRef(base_iri + 'FinancialAsset')
mutualFund = URIRef(base_iri + 'MutualFund')
stock = URIRef(base_iri + 'Stock')
etf = URIRef(base_iri + 'ETF')
sector = URIRef(base_iri + 'Sector')
typeClass = URIRef(base_iri + 'Type')

#object properties
hasLowAndHighPrice = URIRef(base_iri + 'hasLowAndHighPrice')
hasCurrency = URIRef(base_iri + 'hasCurrency')
hasExhange = URIRef(base_iri + 'hasExhange')
hasSectorName = URIRef(base_iri + 'hasSectorName')
hasType = URIRef(base_iri + 'hasType')

#data properties
currencyName = URIRef(base_iri + 'currencyName')
yieldETF = URIRef(base_iri + 'yieldETF')
yieldMF = URIRef(base_iri + 'yield')
ytdReturn = URIRef(base_iri + 'ytdReturn')
morningStarRiskRating = URIRef(base_iri + 'morningStarRiskRating')
morningStarOverallRating = URIRef(base_iri + 'morningStarOverallRating')
navPrice = URIRef(base_iri + 'NAV')
dayLow = URIRef(base_iri + 'dayLowPrice')
dayHigh = URIRef(base_iri + 'dayHighPrice')
trailingEps = URIRef(base_iri + 'Eps')
trailingPE = URIRef(base_iri + 'PE')
tickerSymbol = URIRef(base_iri + 'tickerSymbol')
exchangeName = URIRef(base_iri + 'exchangeName')
exchangeTimezoneName = URIRef(base_iri + 'exchangeTimezoneName')
exchangeTimezoneShortName = URIRef(base_iri + 'exchangeTimezoneShortName')
sectorName = URIRef(base_iri + 'sectorName')
twoHundredDayAverage = URIRef(base_iri + 'twoHundredDayAverage')
fiftyDayAverage = URIRef(base_iri + 'fiftyDayAverage')
typeName = URIRef(base_iri + 'typeName')
```

```python

for item in result_data:
  item_iri = URIRef(base_iri + item['symbol'])

  graph.add((item_iri, RDF.type, financialAsset))
  graph.add((item_iri, dayHigh, Literal(item['dayHigh'])))
  graph.add((item_iri, dayLow, Literal(item['dayLow'])))
  graph.add((item_iri, fiftyDayAverage, Literal(item['fiftyDayAverage'])))
  graph.add((item_iri, twoHundredDayAverage, Literal(item['twoHundredDayAverage'])))

  if item['type'] == 'ETF':    
    graph.add((item_iri, RDF.type, etf))
    graph.add((item_iri, navPrice, Literal(item['navPrice'])))
    graph.add((item_iri, yieldETF, Literal(item['yield'])))
  elif item['type'] == 'MUTUALFUND':
    graph.add((item_iri, RDF.type, mutualFund))
    graph.add((item_iri, morningStarRiskRating, Literal(item['morningStarRiskRating'])))
    graph.add((item_iri, morningStarOverallRating, Literal(item['morningStarOverallRating'])))
    graph.add((item_iri, yieldMF, Literal(item['yield'])))
    graph.add((item_iri, ytdReturn, Literal(item['ytdReturn'])))
  elif item['type'] == 'EQUITY':
    graph.add((item_iri, RDF.type, mutualFund))
    graph.add((item_iri, trailingEps, Literal(item['trailingEps'])))
    graph.add((item_iri, trailingPE, Literal(item['trailingPE'])))
  else:
    continue

  currency_iri = URIRef(base_iri + 'currency' + item['symbol'])
  graph.add((currency_iri, RDF.type, currency))
  graph.add((currency_iri, currencyName, Literal(item['currency'])))
  graph.add((item_iri, hasCurrency, currency_iri))

  type_iri = URIRef(base_iri + 'type' + item['symbol'])
  graph.add((type_iri, RDF.type, typeClass))
  graph.add((type_iri, typeName, Literal(item['type'])))
  graph.add((item_iri, hasType, type_iri))

  exchange_iri = URIRef(base_iri + 'exchange' + item['symbol'])
  graph.add((exchange_iri, RDF.type, exchange))
  graph.add((exchange_iri, exchangeName, Literal(item['exchange'])))
  graph.add((exchange_iri, exchangeTimezoneName, Literal(item['exchangeTimezoneName'])))
  graph.add((exchange_iri, exchangeTimezoneShortName, Literal(item['exchangeTimezoneShortName'])))
  graph.add((item_iri, hasExhange, exchange_iri))

  if item['dayLow'] and item['dayHigh']:
    has_low_and_high_price_iri = URIRef(base_iri + 'hasLowAndHighPrice' + item['symbol'])
    graph.add((has_low_and_high_price_iri, RDF.type, lowAndHighPrice))
    graph.add((has_low_and_high_price_iri, dayLow, Literal(item['dayLow'])))
    graph.add((has_low_and_high_price_iri, dayHigh, Literal(item['dayHigh'])))
    graph.add((item_iri, hasLowAndHighPrice, has_low_and_high_price_iri))

  if 'sector' in item:
    sector_iri = URIRef(base_iri + 'sector' + item['symbol'])
    graph.add((sector_iri, RDF.type, sector))
    graph.add((sector_iri, sectorName, Literal(item['sector'])))
    graph.add((item_iri, hasSectorName, sector_iri))


print(graph.serialize())
```

## Реализуем sparql запросы

Вывести 10 американских акций с самым высоким коэфициентом P/E

```python
g = graph.query(''' 
PREFIX : <http://www.semanticweb.org/anton/ontologies/2021/3/economics#>
SELECT ?FinancialAsset ?PE ?CurrencyName
WHERE {
  ?FinancialAsset :PE ?PE;
  :hasCurrency ?Currency;
  {
    SELECT ?CurrencyName
    WHERE {
      ?Currency a :Currency;
      :currencyName ?CurrencyName;
      FILTER (?CurrencyName = "USD")
    }
  }
  FILTER (
    ?PE >= "50"^^xsd:double && ?PE != "None"
  )
} ORDER BY DESC(?PE)
LIMIT 10
''')

for row in g:
  print(row[0].split('#')[1], row[1], row[2])
```

Вывести фонды где разница между отретием и закрытие больше 1$

```python
g = graph.query(''' 
PREFIX : <http://www.semanticweb.org/anton/ontologies/2021/3/economics#>
SELECT ?FinancialAsset ?range
WHERE {
   ?FinancialAsset :dayLowPrice ?dayLowPrice;
   :dayHighPrice ?dayHighPrice;
   :hasCurrency ?Currency;
   :hasType ?Type;
   {
      SELECT ?typeName
      WHERE {
        ?Type a :Type;
        :typeName ?typeName;
        FILTER (?typeName = 'ETF')
      }
    }
    BIND((?dayHighPrice - ?dayLowPrice) AS ?range)
    FILTER (?range >= 1.0)
}
LIMIT 10
''')

for row in g:
  print(row[0].split('#')[1], str(row[1]))
```

```python
import collections

distinctForPartitions = True
res = Graph()

def generateVoID(g):
    typeMap = collections.defaultdict(set)
    classes = collections.defaultdict(set)
    for e, c in g.subject_objects(RDF.type):
        classes[c].add(e)
        typeMap[e].add(c)

    triples = 0
    subjects = set()
    objects = set()
    properties = set()
    classCount = collections.defaultdict(int)
    propCount = collections.defaultdict(int)

    classProps = collections.defaultdict(set)
    classObjects = collections.defaultdict(set)
    propSubjects = collections.defaultdict(set)
    propObjects = collections.defaultdict(set)

    for s, p, o in g:

        triples += 1
        subjects.add(s)
        properties.add(p)
        objects.add(o)

        # class partitions
        if s in typeMap:
            for c in typeMap[s]:
                classCount[c] += 1
                if distinctForPartitions:
                    classObjects[c].add(o)
                    classProps[c].add(p)

        # property partitions
        propCount[p] += 1
        if distinctForPartitions:
            propObjects[p].add(o)
            propSubjects[p].add(s)

    dataset = URIRef("http://www.semanticweb.org/asususer/ontologies/2021/3/untitled-ontology-15")


    res.add((dataset, RDF.type, VOID.Dataset))

    # basic stats
    res.add((dataset, VOID.triples, Literal(triples)))
    res.add((dataset, VOID.classes, Literal(len(classes))))

    res.add((dataset, VOID.distinctObjects, Literal(len(objects))))
    res.add((dataset, VOID.distinctSubjects, Literal(len(subjects))))
    res.add((dataset, VOID.properties, Literal(len(properties))))

    for i, c in enumerate(classes):
        part = URIRef(dataset + "_class%d" % i)
        res.add((dataset, VOID.classPartition, part))
        res.add((part, RDF.type, VOID.Dataset))

        res.add((part, VOID.triples, Literal(classCount[c])))
        res.add((part, VOID.classes, Literal(1)))

        res.add((part, VOID["class"], c))

        res.add((part, VOID.entities, Literal(len(classes[c]))))
        res.add((part, VOID.distinctSubjects, Literal(len(classes[c]))))

        if distinctForPartitions:
            res.add(
                (part, VOID.properties, Literal(len(classProps[c]))))
            res.add((part, VOID.distinctObjects,
                    Literal(len(classObjects[c]))))

    for i, p in enumerate(properties):
        part = URIRef(dataset + "_property%d" % i)
        res.add((dataset, VOID.propertyPartition, part))
        res.add((part, RDF.type, VOID.Dataset))

        res.add((part, VOID.triples, Literal(propCount[p])))
        res.add((part, VOID.properties, Literal(1)))

        res.add((part, VOID.property, p))

        if distinctForPartitions:

            entities = 0
            propClasses = set()
            for s in propSubjects[p]:
                if s in typeMap:
                    entities += 1
                for c in typeMap[s]:
                    propClasses.add(c)

            res.add((part, VOID.entities, Literal(entities)))
            res.add((part, VOID.classes, Literal(len(propClasses))))

            res.add((part, VOID.distinctSubjects,
                    Literal(len(propSubjects[p]))))
            res.add((part, VOID.distinctObjects,
                    Literal(len(propObjects[p]))))
    
    return res
```
