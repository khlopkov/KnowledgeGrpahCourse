```python
import re

def checkPattern(string, text):
    pattern = re.compile(string)
    return len(re.findall(pattern, text)) > 0


def filterContains(df, pattern):
    return df[df['title'].str.lower().str.contains(pattern)]


def filterEndswith(df, pattern):
    return df[df['title'].str.lower().str.endswith(pattern)]
    

def addCategories(df, categories):
    res_df = pd.DataFrame()
    for (pattern, arg) in categories: 
        filtered_df = None
        if arg == 1:
          filtered_df = filterEndswith(df, pattern)
        else:
          filtered_df = filterContains(df, pattern)
        filtered_df['categorie'] = pattern.replace(' ', '_')
        res_df = pd.concat([res_df, filtered_df]) 
    return res_df

def addBrands(df, brands):
    res_df = pd.DataFrame()
    for pattern in brands: 
        filtered_df = filterContains(df, pattern.lower())
        filtered_df['brand'] = pattern.replace(' ', '_').lower()
        res_df = pd.concat([res_df, filtered_df]) 
    return res_df
        
        

new_df = addCategories(df, categories)

newbrand_df = addBrands(new_df, brands)
newbrand_df.iloc[933]
```

```python
def divideFeatures(df):
    local_df = df.copy()
    for i, e in local_df.iterrows():
        divided_features = [[f.strip() for f in feature.lower().split(':')[:2]] for feature in e.topFeatures]
        e.topFeatures = list(filter(lambda f: len(f) == 2, divided_features))
    return local_df

divided_df = divideFeatures(newbrand_df)
divided_df
```

```python
def priceConvert(df):
    local_df = df.copy()
    for i, e in local_df.iterrows():
        e.price =  float(e.price.replace(',', '').split('£')[-1])
    return local_df

price_df = priceConvert(divided_df)

res = []
for i, e in price_df.iterrows():
    # res.append(e.topFeatures)
    for f in e.topFeatures:
        res.append(f)

price_df.to_csv('/content/drive/MyDrive/ont/dataframe.csv')
len(dict(res).values())
price_df.iloc[933]
```

```python

from rdflib import Graph, Literal, BNode, Namespace, URIRef
from rdflib.namespace import XSD, RDF, OWL, RDFS, VOID
from urllib.parse import quote_plus
```

```python
g = Graph()
g.parse("/content/drive/MyDrive/ont/ont_blank1.owl", format="turtle")

prefix = 'urn:bike'


brands = list(price_df.brand.unique())

def createUri(v):
    s = v.replace(' ', '_').replace('(', '').replace(')', '').replace('"', "").replace("^", "").replace("\\", "").replace("%", "")
    # .replace("'", "")
    return URIRef(s.encode("ascii", "ignore").decode())

def addCategorie(g, categorie):
    Categorie = createUri(prefix + '#Categorie')   
    uri = createUri(prefix + f'#{categorie}')   
    name = createUri(prefix + '#name') 
    g.add((uri, RDF.type, Categorie))
    g.add((uri, name, Literal(categorie.replace('_', ' '))))
    return uri

def addBrand(g, brand):
    Brand = createUri(prefix + '#Brand')   
    uri = createUri(prefix + f'#{brand}')   
    name = createUri(prefix + '#name') 
    g.add((uri, RDF.type, Brand))
    g.add((uri, name, Literal(brand.replace('_', ' '))))
    return uri

def addFeature(g, feature):
    Feature = createUri(prefix + '#Feature')   
    Name = createUri(prefix + '#Name')   
    Value = createUri(prefix + '#Value')   
    fname = f'#{feature[0] + "_" + feature[1]}'
    uriF = createUri(prefix + fname)   
    uriN = createUri(prefix + f'#{feature[0]}')   
    uriV = createUri(prefix + f'#{feature[1]}')   
    name = createUri(prefix + '#name') 
    val = createUri(prefix + '#val') 

    hasFeature = createUri(prefix + '#hasFeature') 
    hasName = createUri(prefix + '#hasName') 
    hasValue = createUri(prefix + '#hasValue') 

    g.add((uriF, RDF.type, Feature))
    g.add((uriN, RDF.type, Name))
    g.add((uriV, RDF.type, Value))
    g.add((uriN, name, Literal(feature[0])))
    g.add((uriV, val, Literal(feature[1])))
    g.add((uriF, hasName, uriN))
    g.add((uriF, hasValue, uriV))

    return uriF


def addPart(g, series):
    partStr = series.title.lower().replace(' ', '_')
    nameStr = series.title
    priceStr = str(series.price)
    features = series.topFeatures
    categorie = series.categorie
    brand = series.brand

    Part = createUri(prefix + '#Part')   
    uri = createUri(prefix + f'#{partStr}')   
    name = createUri(prefix + '#name') 
    price = createUri(prefix + '#price') 

    g.add((uri, RDF.type, Part))
    g.add((uri, name, Literal(nameStr)))
    g.add((uri, price, Literal(priceStr, datatype=XSD.float)))

    hasCategorie = createUri(prefix + '#hasCategorie') 
    hasBrand = createUri(prefix + '#hasBrand') 
    hasFeature = createUri(prefix + '#hasFeature') 
    contain = createUri(prefix + '#contain') 

    uriB = addBrand(g, brand)
    uriC = addCategorie(g, categorie)
    g.add((uri, hasBrand, uriB))
    g.add((uri, hasCategorie, uriC))
    # g.add((uriB, contain, uri))
    # g.add((uri, contain, uriC))

    for feature in features:
        uriF = addFeature(g, feature)
        g.add((uri, hasFeature, uriF))

for i, s in price_df.iterrows():
    addPart(g, s)


POPULATED_OWL_FILENAME = "/content/drive/MyDrive/ont/ont_populated_full.owl"
g.serialize(destination=POPULATED_OWL_FILENAME, format="turtle")
```
