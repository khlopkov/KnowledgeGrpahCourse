```python
graph = rdflib.Graph()
graph.parse('/content/gdrive/MyDrive/Графы/games.owl')
```

```python
prefix = r'http://www.semanticweb.org/naymoll/ontologies/2022/3/games.owl'
```

Classes

```python
from rdflib import URIRef

class Classes:
  PRODUCT  = URIRef(prefix + '#Product')
  DLC      = URIRef(prefix + '#DLC')
  GAME     = URIRef(prefix + '#Game')
  OS       = URIRef(prefix + '#OS')
  GENRE    = URIRef(prefix + '#Genre')
  LANGUAGE = URIRef(prefix + '#Language')
  REVIEW   = URIRef(prefix + '#Review')
  TAG      = URIRef(prefix + '#Tag')
```

Object property

```python
class ObjectProperty:
  COMPATIBLE_WITH   = URIRef(prefix + '/product#compatible_with')
  HAS_GENRE         = URIRef(prefix + '/product#has_genre')
  HAS_LANG_SUPPORT  = URIRef(prefix + '/product#has_lang_support')
  HAS_REVIEW        = URIRef(prefix + '/product#has_review')
  HAS_TAG           = URIRef(prefix + '/product#has_tag')
  DLC               = URIRef(prefix + '/product#dlc')
  HAS_DLC           = URIRef(prefix + '/game#has_dlc')
  IS_DLC_OF         = URIRef(prefix + '/dlc#is_dlc_of')
```

Product data property

```python
class ProductDP:
  ID           = URIRef(prefix + '/product#id')
  RELEASE_DATE = URIRef(prefix + '/product#release_date')
  COOL_DESC    = URIRef(prefix + '/product#cool_desc')
  LEAD_DESC    = URIRef(prefix + '/product#lead_desc')
  FULL_DESC    = URIRef(prefix + '/product#full_desc')
  TITLE        = URIRef(prefix + '/product#title')
  CHANGELOG    = URIRef(prefix + '/product#changelog')
  IS_PRE_ORDER = URIRef(prefix + '/product#is_pre_order')
  SLUG         = URIRef(prefix + '/product#slug')
  UNTIL        = URIRef(prefix + '/product#until')
  ACTIVE       = URIRef(prefix + '/product#active')
```

Review data property

```python
class ReviewDP:
  ID          = URIRef(prefix + '/review#id')
  RATING      = URIRef(prefix + '/review#rating')
  TITLE       = URIRef(prefix + '/review#title')
  DESCRIPTION = URIRef(prefix + '/review#description')
  DOWNVOTES   = URIRef(prefix + '/review#downvotes')
  UPVOTES     = URIRef(prefix + '/review#upvotes')
  DATE        = URIRef(prefix + '/review#date')
```

```python
class OS:
  LINUX   = URIRef(prefix + '/os/linux')
  MAC_OS  = URIRef(prefix + '/os/mac_os')
  WINDOWS = URIRef(prefix + '/os/windows')
```

```python
from rdflib import Literal
from rdflib.namespace import RDF, XSD
```

```python
def add_tags(product_uri, tags):
  for tag in tags:
    tag_uri = URIRef(prefix + f'/tag/{tag}')
    
    if (tag_uri, None, None) not in graph:
      graph.add((tag_uri, RDF.type, Classes.TAG))
      
    graph.add((product_uri, ObjectProperty.HAS_TAG, tag_uri))
```

```python
def add_genres(product_uri, genres):
  for genre in genres:
    genre_uri = URIRef(prefix + f'/genre/{genre}')
    
    if (genre_uri, None, None) not in graph:
      graph.add((genre_uri, RDF.type, Classes.GENRE))
      
    graph.add((product_uri, ObjectProperty.HAS_GENRE, genre_uri))
```

```python
def add_languages(product_uri, languages):
  for language in languages:
    language_uri = URIRef(prefix + f'/language/{language}')
    
    if (language_uri, None, None) not in graph:
      graph.add((language_uri, RDF.type, Classes.LANGUAGE))
      
    graph.add((product_uri, ObjectProperty.HAS_LANG_SUPPORT, language_uri))
```

```python
def add_dlcs(product_uri, dlcs):
  for dlc in dlcs:
    dlc_uri = URIRef(prefix + f'/dlc/{dlc}')  
    graph.add((product_uri, ObjectProperty.HAS_DLC, dlc_uri))
```

```python
def add_os_support(uri, product):
  if product.windows:
    graph.add((uri, ObjectProperty.COMPATIBLE_WITH, OS.WINDOWS))
  if product.linux:
    graph.add((uri, ObjectProperty.COMPATIBLE_WITH, OS.LINUX))
  if product.osx:
    graph.add((uri, ObjectProperty.COMPATIBLE_WITH, OS.MAC_OS))
```

```python
def add_description(uri, product):
  if pd.notnull(product.lead):
    graph.add((uri, ProductDP.LEAD_DESC, Literal(product.lead)))
  if pd.notnull(product.full):
    graph.add((uri, ProductDP.FULL_DESC, Literal(product.full)))
  if pd.notnull(product.whats_cool_about_it):
    graph.add((uri, ProductDP.COOL_DESC, Literal(product.whats_cool_about_it)))
```

```python
def add_reviews(uri, reviews):
  for _, review in reviews.iterrows():
    id = review.id
    review_uri = URIRef(prefix + f'/review/{id}')
    
    graph.add((review_uri, RDF.type,             Classes.REVIEW))
    graph.add((review_uri, ReviewDP.ID,          Literal(id)))
    graph.add((review_uri, ReviewDP.RATING,      Literal(review.rating)))
    graph.add((review_uri, ReviewDP.TITLE,       Literal(review.title)))
    graph.add((review_uri, ReviewDP.DESCRIPTION, Literal(review.description)))
    graph.add((review_uri, ReviewDP.DOWNVOTES,   Literal(review.downvotes)))
    graph.add((review_uri, ReviewDP.UPVOTES,     Literal(review.upvotes)))
    graph.add((review_uri, ReviewDP.DATE,        Literal(review.date)))

    graph.add((uri, ObjectProperty.HAS_REVIEW, review_uri))
```

```python
# title slug languages game_type is_pre_order release_date dlcs changelog windows osx linux lead full whats_cool_about_it active until genres tags

def add_product(id, product):
  if product.game_type == 'game':
    uri = URIRef(prefix + f'/game/{id}')
    graph.add((uri, RDF.type, Classes.GAME))
  else:
    uri = URIRef(prefix + f'/dlc/{id}')
    graph.add((uri, RDF.type, Classes.DLC))

  graph.add((uri, ProductDP.ID,           Literal(id)))
  graph.add((uri, ProductDP.TITLE,        Literal(product.title)))
  graph.add((uri, ProductDP.SLUG,         Literal(product.slug)))
  graph.add((uri, ProductDP.IS_PRE_ORDER, Literal(product.is_pre_order)))

  if pd.notnull(product.release_date):
    graph.add((uri, ProductDP.RELEASE_DATE, Literal(product.release_date, datatype=XSD.dateTime)))

  add_languages(uri, product.languages)
  add_os_support(uri, product)
  add_description(uri, product)
  
  if pd.notnull(product.changelog):
    graph.add((uri, ProductDP.CHANGELOG, Literal(product.changelog)))

  graph.add((uri, ProductDP.ACTIVE, Literal(product.active)))
  if pd.notnull(product.until):
    graph.add((uri, ProductDP.UNTIL, Literal(product.until)))

  if product.tags:
    add_tags(uri, product.tags)

  if product.genres:
    add_genres(uri, product.genres)

  if product.dlcs:
    add_dlcs(uri, product.dlcs)

  product_reviews = REVIEWS[REVIEWS.product_id == id]
  add_reviews(uri, product_reviews)
```

###FULL

```python
GAMES = GAMES_INFO[GAMES_INFO.game_type == 'game']
DLCS = GAMES_INFO[GAMES_INFO.game_type == 'dlc']

for id, product in GAMES.iterrows():
  add_product(id, product)

for id, product in DLCS.iterrows():
  add_product(id, product)

graph.serialize(destination='/content/gdrive/MyDrive/Графы/filled_games_full.owl', format='n3')
```
