```python
from ampligraph.utils import restore_model

model = restore_model(model_name_path='/content/gdrive/MyDrive/Графы/games.pkl')
```

```python
import pandas as pd

REVIEWS = pd.read_json('/content/gdrive/MyDrive/Графы/reviews.json')
```

```python
REVIEWS
```

```python
ids = pd.read_json('/content/gdrive/MyDrive/Графы/review_ids_cut.json')
```

```python
ids = ids.reset_index().drop(['index'], axis=1)
ids.columns = ['id']
```

```python
merged = pd.merge(REVIEWS, ids, how='left', on=['id'], indicator='exist')
```

```python
filtred = merged[merged.exist == 'both'].drop(['exist'], axis=1)
```

```python
filtred
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(filtred, test_size = 0.2)
```

```python
def get_targets(df):
  X = model.get_embeddings(df.id.values)
  Y = df.rating.values
  return X, Y
```

```python
X_train, Y_train = get_targets(train_df)
X_test, Y_test = get_targets(test_df)
```

```python
import numpy as np

np.isnan(X_test).sum()/X_test.shape[1]
```

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier

clf_model = XGBClassifier(n_estimators=500, max_depth=5, objective="multi:softmax")
```

```python
clf_model.fit(X_train, Y_train)
```

```python
filtred.rating.value_counts(normalize=True)
```

```python
from sklearn import metrics

metrics.accuracy_score(Y_test, clf_model.predict(X_test))
```
