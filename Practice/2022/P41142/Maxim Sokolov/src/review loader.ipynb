```python
from pandas._libs import index
def get_review(product):
  base_url = f'https://reviews.gog.com/v1/products/{product}/reviews'
  page = 1
  reviews = pd.DataFrame()
  
  while True:
    page_url = f'{base_url}?page={page}'
    response = session.get(page_url)
    gog_response = response.json()

    items = gog_response['_embedded']['items']
    for item in items:      
      id = item['id']
      productId = item['productId']
      rating = item['rating']['value']
      title = item['content']['title']
      description = item['content']['description']
      downvotes = item['votes']['downvotes']
      upvotes = item['votes']['upvotes']
      date = item['creationDate']

      data = {'id':id, 'product_id':productId, 'rating':rating, 'title':title, 'description':description, 'downvotes':downvotes, 'upvotes':upvotes, 'date':date}
      df = pd.DataFrame.from_dict([data])
      reviews = reviews.append(df)

    if page >= gog_response['pages']:
      return reviews.reset_index().drop(['index'], axis=1)

    page += 1
```

```python
from dateutil import parser

def remove_timezone(datetime):
  datetime = parser.parse(datetime).replace(tzinfo=None)
  return datetime.isoformat()
```

```python
reviews.date = reviews.date.map(remove_timezone)
```
