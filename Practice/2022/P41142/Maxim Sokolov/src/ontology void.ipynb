```python
import rdflib

graph = rdflib.Graph()
graph.parse('/content/gdrive/MyDrive/Графы/filled_games_cut.owl')
```

```python
from rdflib import URIRef
from rdflib.void import generateVoID

void_graph, d = generateVoID(graph, dataset=URIRef('http://www.semanticweb.org/naymoll/ontologies/2022/3/games.owl/dataset'))
void_graph.serialize(destination='/content/gdrive/MyDrive/Графы/filled_games_cut_void.owl', format='n3')
```
