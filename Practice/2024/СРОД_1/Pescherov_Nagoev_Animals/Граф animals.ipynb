## **Заполнение**

```python
# Чтение данных из CSV файлов
sounds_df = pd.read_csv('sounds.csv')
images_df = pd.read_csv('images.csv')
descriptions_df = pd.read_csv('descriptions.csv')
compositions_df = pd.read_csv('compositions.csv')
```

```python
# Создание графа
g = Graph()

# Пространства имен (namespace)
NS = Namespace("http://www.semanticweb.org/аскер/ontologies/2024/5/animals#")

# Загрузка существующей онтологии в граф
g.parse('animals.rdf')

# Привязка namespace
g.bind("animals", NS)
```

```python
# Добавление данных из sounds.csv
for _, row in sounds_df.iterrows():
    animal = URIRef(NS + row['Уровень 1'])
    sound = URIRef(NS + row['Уровень 2'])
    spectrogram = URIRef(NS + row['Спектрограмма'])

    g.add((animal, NS['иметь_звук'], sound))
    g.add((sound, RDF.type, NS['Аудио']))
    g.add((sound, NS['имеет_спектрограмму'], spectrogram))
    g.add((spectrogram, RDF.type, NS['Спектрограмма']))
```

```python
# Добавление данных из images.csv
for _, row in images_df.iterrows():
    animal = URIRef(NS + row['Уровень 1'])
    image = URIRef(NS + row['Уровень 2'])
    annotated_image = URIRef(NS + row['Размеченный'])

    g.add((animal, NS['иметь_изображение'], image))
    g.add((image, RDF.type, NS['Изображение']))
    g.add((image, NS['имеет_размеченное_изображение'], annotated_image))
    g.add((annotated_image, RDF.type, NS['Размеченное_изображение']))
```

```python
# Добавление данных из descriptions.csv
for _, row in descriptions_df.iterrows():
    animal = URIRef(NS + row['Уровень 1'])
    description = URIRef(NS + row['Уровень 2'])
    keywords = Literal(row['Уровень 3'])
    vector = URIRef(NS + row['Вектор'])

    g.add((animal, NS['иметь_текстовое_описание'], description))
    g.add((description, RDF.type, NS['Текстовое_описание']))
    g.add((description, NS['ключевые_слова'], keywords))
    g.add((description, NS['имеет_векторное_описание'], vector))
    g.add((vector, RDF.type, NS['Вектор_описания']))
```

```python
# Добавление данных из compositions.csv
for _, row in compositions_df.iterrows():
    animal = URIRef(NS + row['Уровень 1'])
    composition = URIRef(NS + row['Уровень 2'])
    locations = Literal(row['Уровень 3'])
    characters = Literal(row['Уровень 4'])
    vector = URIRef(NS + row['Вектор'])

    g.add((animal, NS['иметь_текстовое_произведение'], composition))
    g.add((composition, RDF.type, NS['Текстовое_произведение']))
    g.add((composition, NS['локации'], locations))
    g.add((composition, NS['герои'], characters))
    g.add((composition, NS['имеет_векторное_описание'], vector))
    g.add((vector, RDF.type, NS['Вектор_произведения']))
```

```python
# Сохранение графа в файл в формате Turtle
g.serialize(destination='ontology_output.ttl', format='turtle')
```

```python
from rdflib import Graph, RDF, RDFS

# Создание графа и загрузка данных
g = Graph()
g.parse('ontology_output.ttl', format='turtle')

# Запрос для получения подклассов класса "Животное"
query = """
SELECT ?subclass
WHERE {
    ?subclass rdfs:subClassOf <http://www.semanticweb.org/аскер/ontologies/2024/5/animals#Животное> .
}
"""

# Функция для выполнения SPARQL запроса
def run_query(query):
    result = g.query(query)
    for row in result:
        print(row)

# Выполнение запроса
print("Подклассы Животных:")
run_query(query)
```

```python
from rdflib import Graph, URIRef
from rdflib.plugins.sparql import prepareQuery

# Создаем граф и загружаем данные
g = Graph()
g.parse('ontology_output.ttl', format='turtle')

# SPARQL-запрос для поиска животных с ключевыми словами, содержащими "шерсть"
query = prepareQuery('''
    SELECT DISTINCT ?description
    WHERE {
        ?description <http://www.semanticweb.org/аскер/ontologies/2024/5/animals#ключевые_слова> ?keywords .
        FILTER regex(?keywords, "шерсть", "i")
    }
''')

# Выполнение запроса и вывод результатов
print("Животные с ключевыми словами, содержащими 'шерсть':")
for row in g.query(query):
    print(row)
```

```python
from rdflib import Graph, Namespace, URIRef
from rdflib.plugins.sparql import prepareQuery

# Создаем объект графа RDF и загружаем данные
g = Graph()
g.parse('ontology_output.ttl', format='turtle')

# Определяем префиксы пространства имен
rdf = Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
onto = Namespace("http://www.semanticweb.org/аскер/ontologies/2024/5/animals#")

# Формулируем SPARQL-запрос
query = prepareQuery("""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX onto: <http://www.semanticweb.org/аскер/ontologies/2024/5/animals#>

    SELECT ?composition
    WHERE {
        ?composition onto:герои ?heroes .
        FILTER regex(?heroes, "мышь", "i") .
    }
""", initNs={"rdf": rdf, "onto": onto})

# Выполняем запрос и выводим результаты
results = g.query(query)

print("Произведения с героем 'мышь':")
for row in results:
    print(row.composition)
```
