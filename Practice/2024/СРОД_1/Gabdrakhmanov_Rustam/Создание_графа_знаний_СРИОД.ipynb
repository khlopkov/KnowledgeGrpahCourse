```python
ontology_path = '/content/urn_webprotege_ontology_7680d888-1c2e-48c2-9b9a-aad5aa5d1592.owl'

# Создание графа
g = Graph()

# Загрузка онтологии в граф
g.parse(ontology_path)

# Сохранение графа в файл в формате Turtle
g.serialize(destination='ontology_output.ttl', format='turtle')

# Вывод графа в формате Turtle на экран
print(g.serialize(format='turtle'))
```

```python
ex = Namespace("http://example.org/")
g.bind("ex", ex)

# Добавление данных в онтологию

# Примеры персонажей
Zorro = URIRef(ex.Zorro)
Sanji = URIRef(ex.Sanji)

# Примеры оружия
leg = URIRef(ex.Leg)
katana = URIRef(ex.Katana)

# Примеры магии и техник
water = URIRef(ex.Water)
fire = URIRef(ex.Fire)

sanji_speech = URIRef(ex.Sanji_Speech)
zorro_speech = URIRef(ex.Zorro_Speech)
sward_sound = URIRef(ex.Sward_Sound)
leg_sound = URIRef(ex.Leg_Sound)
```

```python
# Добавление триплетов
g.add((Zorro, RDF.type, ex.characters))
g.add((Zorro, ex.ch_description, Literal("Zorro is a man who wants to become the greatest swordsman")))
g.add((Zorro, ex.has_weapon, katana))


g.add((Sanji, RDF.type, ex.characters))
g.add((Sanji, ex.ch_description, Literal("Sanji is a chef who wants to find ALL blue.")))
g.add((Sanji, ex.has_weapon, leg))
g.add((Sanji, ex.use_magic, fire))

g.add((katana, RDF.type, ex.melee_weapon))
g.add((katana, ex.weapon_has_appearence, Literal("/Photo/Zorro.jpg, /Photo/hero_marking.csv")))

g.add((leg, RDF.type, ex.melee_weapon))
g.add((leg, ex.weapon_has_appearence, Literal("/Photo/Sunji.jpg, /Photo/hero_marking.csv")))

g.add((water, RDF.type, ex.magic))
g.add((water, RDFS.label, Literal("Water fist")))

g.add((fire, RDF.type, ex.magic))
g.add((fire, RDFS.label, Literal("fire punch")))

# Добавление триплетов для речи
g.add((sanji_speech, RDF.type, ex.speach))
g.add((sanji_speech, ex.has_speech, Literal("/Text/1062_sab_prepared.xlsx, Style: Санджи")))

g.add((zorro_speech, RDF.type, ex.speach))
g.add((zorro_speech, ex.has_speech, Literal("/Text/1062_sab_prepared.xlsx, Style: Зорро")))


# Добавление триплетов для звуков
g.add((leg, RDF.type, ex.sound))
g.add((leg, ex.has_sound, Literal("/Audio/1062_audio.csv, CharacterName: Санджи")))

g.add((katana, RDF.type, ex.sound))
g.add((katana, ex.has_sound, Literal("/Audio/1062_audio.csv, CharacterName: Зорро")))

g.add((sward_sound, RDF.type, ex.sound))
g.add((sward_sound, ex.sound_description, Literal("/Audio/1062_audio.csv, CharacterName: Санджи", "Description")))

g.add((leg_sound, RDF.type, ex.sound))
g.add((leg_sound, ex.sound_description, Literal("/Audio/1062_audio.csv, CharacterName: Зорро", "Description")))
```

```python
g.serialize(destination="anime_ontology.rdf", format="turtle")

# Вывод графа на экран
print(g.serialize(format="turtle"))
```

```python

ex = Namespace("http://example.org/")
webprotege = Namespace("http://webprotege.stanford.edu/")

query = """
PREFIX ex: <http://example.org/>
PREFIX wp: <http://webprotege.stanford.edu/>

SELECT (COUNT(DISTINCT ?melee_weapon) AS ?count)
WHERE {
  ?melee_weapon a ex:melee_weapon .
}
"""

# Выполнение запроса
results = g.query(query)

# Вывод результата
for row in results:
    print(row)
    print(f"Количество героев: {row['count']}")
```

```python
# Запрос SPARQL для получения количества героев с оружием ближнего боя
query = """
PREFIX ex: <http://example.org/>

SELECT (COUNT(DISTINCT ?hero) AS ?count)
WHERE {
  ?hero a ex:characters ;
        ex:has_weapon ?weapon .
  ?weapon a ex:melee_weapon .
}
"""

# Выполнение запроса
results = g.query(query)

# Вывод результата
for row in results:
    print(f"Количество героев с оружием ближнего боя: {row['count']}")
```

```python
query = """
PREFIX ex: <http://example.org/>
PREFIX wp: <http://webprotege.stanford.edu/>

SELECT ?hero ?weapon ?appearance
WHERE {
  ?hero a ex:characters ;
        ex:has_weapon ?weapon .
  ?weapon ex:weapon_has_appearence ?appearance .
}
"""

# Выполнение запроса
results = g.query(query)

# Вывод результата
for row in results:
    print(f"Герой: {row['hero']}, Оружие: {row['weapon']}, Внешний вид оружия: {row['appearance']}")
```

```python
query = """
PREFIX ex: <http://example.org/>

SELECT ?hero ?weapon ?sound
WHERE {
  ?hero a ex:characters ;
        ex:use_magic ?magic ;
        ex:has_weapon ?weapon .
  ?weapon ex:has_sound ?sound .
}
"""

# Выполнение запроса
results = g.query(query)

# Вывод результата
for row in results:
    print(f"Герой: {row['hero']}, Оружие: {row['weapon']}, Звук: {row['sound']}")
```
