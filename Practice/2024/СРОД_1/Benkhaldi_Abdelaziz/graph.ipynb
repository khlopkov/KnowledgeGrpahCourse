```python
query = """
PREFIX ex: <http://example.org/auto#>
SELECT ?segment ?noise
WHERE {
    ?segment rdf:type ex:AudioSegment .
    ?segment ex:describesNoise ?noise .
}
"""
for row in g.query(query):
    print(f"Segment: {row.segment}, Noise: {row.noise}")
```

```python
query = """
PREFIX ex: <http://example.org/auto#>
SELECT ?image ?annotation ?part ?symptom
WHERE {
    ?image rdf:type ex:Image .
    ?image ex:hasAnnotation ?annotation .
    ?annotation ex:highlightsPart ?part .
    ?annotation ex:indicatesError ?symptom .
}
"""

for row in g.query(query):
    print(f"Image: {row.image}, Annotation: {row.annotation}, Part: {row.part}, Symptom: {row.symptom}")
```
